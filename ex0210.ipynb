```python
#newsapi 주소를 가져와라
res=requests.get(
'https://newsapi.org/v2/top-headlines?country=kr&apiKey=9f5baf7d9f3f42879a20d7d19d9886e4'
)
# 가져온 뉴스를 제이슨타입으로 바꾸고 result라는 변수에 담아라
result = res.json()


data = result.get('articles')


for i in data:
    y = i.get('publishedAt')[:4]
    m = i.get('publishedAt')[5:7]
    d = i.get('publishedAt')[8:10]
    s = i.get('publishedAt')[11:13]
    t = i.get('publishedAt')[14:16]
    print(f"{y}년 {m}월 {d}일 {s}시 {t}분")
```
